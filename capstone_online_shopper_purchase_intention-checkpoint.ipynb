```python
# Clustering Administrative duration vs Bounce Rate
```

```python
x = df.iloc[:, [1, 6]].values
x.shape
```

```python
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i,
              init = 'k-means++',
              max_iter = 300,
              n_init = 10,
              random_state = 0,
              algorithm = 'elkan',
              tol = 0.001)
    km.fit(x)
    labels = km.labels_
    wcss.append(km.inertia_)
    
plt.rcParams['figure.figsize'] = (15, 7)
plt.plot(range(1, 11), wcss)
plt.grid()
plt.tight_layout()
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)

plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'pink', label = 'Un-interested Customers')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'yellow', label = 'General Customers')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'cyan', label = 'Target Customers')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'centeroid')

plt.title('Administrative Duration vs Bounce Rate', fontsize = 20)
plt.grid()
plt.xlabel('Administrative Duration')
plt.ylabel('Bounce Rates')
plt.legend()
plt.show()
```

```python
# Oversampling of Data to avoid Class imbalance ( USING SMOTE)
```

```python
smt = SMOTE()
X_sm, y_sm = smt.fit_sample(X,y)
np.bincount(y_sm)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm,y_sm,train_size=0.7,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
lr= LogisticRegression(random_state=0)
lr.fit(X_train, y_train)
LR_y_pred = lr.predict(X_test)

lr_test_accuracy=metrics.accuracy_score(y_test,LR_y_pred)
lr_train_accuracy=lr.score(X_train, y_train)
lr_test_precision = metrics.precision_score(y_test,LR_y_pred)

print("Test Accuracy for Logistic Regression:\t",lr_test_accuracy)
print()
print("Training Accuracy for Logistic Regression:\t",lr_train_accuracy)
print("\nConfusion Matirx:\n\n",metrics.confusion_matrix(y_test,LR_y_pred))
print()
print("\nPrecision Score:\n\n",lr_test_precision)
```

```python
from sklearn import metrics
def draw_roc(actual,probs):
    fpr,tpr,thresholds = metrics.roc_curve(actual,probs,drop_intermediate=False)
    auc_score=metrics.roc_auc_score(actual,probs)
    plt.figure(figsize=(6,6))
    plt.plot(fpr,tpr,label='ROC curve(area = %0.2f)' %auc_score)
    plt.plot([0,1],[0,1],'k--')
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.05])
    plt.xlabel('False positve rate')
    plt.ylabel('True positve rate')
    plt.title('ROC CURVE')
    plt.legend(loc="lower right")
    plt.show()
    
    return fpr,tpr,thresholds
```

```python
draw_roc(y_test,LR_y_pred)
```
